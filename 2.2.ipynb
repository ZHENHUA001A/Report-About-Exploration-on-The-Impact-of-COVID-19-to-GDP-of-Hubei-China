```python
# Convert the total GDP as of the quarter into the increment of GDP for the quarter
L1 = GDP1
L1
```

```python
L2 = []
for i in range(4):
    L2.append(GDP2[i] - GDP1[i])
L2
```

```python
L3 = []
for i in range(4):
    L3.append(GDP3[i] - GDP2[i])
L3
```

```python
L4 = []
for i in range(4):
    L4.append(GDP4[i] - GDP3[i])
L4
```

```python
L = []
for i in range(4):
    L.append(L1[i])
    L.append(L2[i])
    L.append(L3[i])
    L.append(L4[i])
L
```

```python
r = []
for i in range(1,len(L) - 1):
    r.append((L[i]- L[i - 1])/L[i - 1])
r
```

```python
r_lable = []
r_lable.append('2018_2')
y = 2019
for i in range(len(r) - 1):
    if i % 4 == 2:
        r_lable.append(str(y) + '_1')
        y = y + 1
    else:
        r_lable.append(' ')
r_lable
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
 
GDP1 = [8188,9110,6379,9872]
GDP2 = [17958,19895,17480,22777]
GDP3 = [27634,30509,29779,34731]
GDP4 = [39366,45828,43443,34731]

data=[GDP1,GDP2,GDP3,GDP4]
data1=[]
for i in range(4):
    l = []
    l.append(GDP1[i])
    l.append(GDP2[i])
    l.append(GDP3[i])
    l.append(GDP4[i])
    data1.append(l)
    


def ladar(data,feature):
    N = 4  # number of 
    angles=np.linspace(0, 2*np.pi, N, endpoint=False) # set angle
    angles=np.concatenate((angles, [angles[0]])) # Make the radar map closed


    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True) # Set polar coordinates
    sam = ['r-', 'm-', 'g-', 'b-', 'y-', 'k-', 'w-','c-'] # type
    color = ['black','b','y','#FFB6C1']
    lab = [1,2,3,4] # laber

    feature=np.concatenate((feature,[feature[0]]))
    for i in range(len(data)):
        values = data[i]
         # names of features

        # Make the radar map closed
        values=np.concatenate((values,[values[0]]))
        


        ax.plot(angles, values, sam[i], linewidth=2) # Draw a line chart
        ax.fill(angles, values,color[i], alpha=0.5) # fill color
        ax.set_thetagrids(angles * 180/np.pi, feature) # add labers
        ax.set_ylim(auto = True) # set scope
        plt.title('') # add the title
        ax.grid(True) # Add grid lines
        lab.append(str(i+1))
    plt.savefig("ladar.jpg")  # save the picture
    plt.show()
```

```python
feature = ['2018','2019','2020','2021']
ladar(data,feature)
```
